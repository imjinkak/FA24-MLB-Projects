```python
iris_table = Table.read_table("/kaggle/input/iris/Iris.csv")
iris_df= pd.read_csv("/kaggle/input/iris/Iris.csv")
iris_table.show(10)
```

```python
mapping = {"Iris-setosa": 0, "Iris-versicolor":1, "Iris-virginica":2}
y = iris_df["Species"].map(mapping)
X = iris_df.drop(columns = "Species")
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.35, random_state = 42 )
```

```python
knn = KNeighborsClassifier()
linReg = LinearRegression()
svc = svm.LinearSVC()
rcf = RandomForestClassifier()
```

```python
knn.fit(X_train,y_train)
knn_score = knn.score(X_test, y_test)
knn_score
```

```python
linReg.fit(X_train,y_train)
linReg_score = linReg.score(X_test, y_test)
linReg_score
```

```python
svc.fit(X_train,y_train)
svc_score = svc.score(X_test, y_test)
svc_score
```

```python
rcf.fit(X_train,y_train)
rcf_score = rcf.score(X_test, y_test)
rcf_score
```
